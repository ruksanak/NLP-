```python
def preprocess(text):
    result = []
    tweet = re.sub("(@_?[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", text)
    for token in simple_preprocess(tweet):
        if token not in STOPWORDS:
            result.append(lemmatize_stemming(token))
    return result
```

```python
def remove_special_word1(tweet_str):
    tweet = re.sub("(@_?[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet_str).split()
    cleaned_tweet = ' '.join([w for w in tweet if w not in stopwords.words('english')])
    return cleaned_tweet

clean_text = tweets_df2['Text'].apply(lambda x: remove_special_word1(x))
```

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
import torch
```

```python
#Used pretrained model for the score - load it in the notbook
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
def bert_senti_score(tweet):
    tokens = tokenizer.encode(tweet, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))+1
```

```python
tweets_df2['clean_text']  = clean_text

temp = tweets_df2.head(300)
```

```python
temp['bert_scores'] = temp['clean_text'].apply(lambda x: bert_senti_score(x))
```
